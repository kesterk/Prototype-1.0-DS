### A FUNCTION THAT SEARCHES FOR PDF FILES IN A FOLDER AND ETRACT THE TABLES TO EXCEL SHEET

```python
def Table_Extraction(folder, encoding,Name,area=[0,12.75,1000 ,600],lattice=False):
    folder=folder
    paths = [folder +"\\" + fn for fn in os.listdir(folder) if fn.endswith('.pdf')]
    for path in paths:
        df = tabula.read_pdf(path,multiple_tables=True,encoding=encoding, pages='all',area=area)
        i=1
        for table in df:
            table.to_excel(Name+str(i)+'.xlsx', header=True)
            i= i+1
```

```python
Table_Extraction(area=[0, 12.75, 1500, 300],folder=r"C:\Users\USER\Desktop\BCC",encoding='ISO-8859-1',Name='TABLE ')
```
